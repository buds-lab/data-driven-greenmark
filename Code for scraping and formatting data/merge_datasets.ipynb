# This notebook is to merge the old and new datasets

```python
import pandas as pd
```

```python
old = pd.read_csv("bca_greenmark_bdgs_cleaned.csv")
new = pd.read_csv("GM_buildings_formatted.csv")
```

## Since the format of the BCA ID is different between the two datasets, I use the building name and year to differentiate the buildings

```python
pd.concat([old, new])
```

```python
old_bdgs = {}
for _,row in old.iterrows():
    old_bdgs[row["building"], row["fy"]] = row
old_bdgs
```

```python
new_bdgs = []
for _,row in new.iterrows():
    new_bdgs.append([row["building"], row["fy"]])
new_bdgs
```

```python
# Find buildings from old dataset not in new dataset
old_not_in_new = list(filter(lambda x: x not in new_bdgs, old_bdgs.keys()))
```

```python
for i in old_not_in_new:
    new = new.append(old_bdgs[i])
```

```python
new = new.reset_index().drop(columns=["index"])
```

```python
columns = ['bca_id',
 'building',
 'award',
 'fy',
 'gfa',
 'address',
 'postal_code',
 'district',
 'developer',
 'architect',
 'structural_eng',
 'me_eng',
 'land_cons',
 'qs',
 'main_contractor',
 'fac_mgt',
 'esd_cons',
 'esco_cons',
 'green_cons',
 'description',
 'acoustics_cons',
 'facade_cons',
 'light_cons',
 'green_features']
```

```python
new = new[columns]
```

```python
new.to_csv("GM_buildings_merged.csv", index=False)
```
