```python
# Building with 5 stakeholders; create a copy with the 3 columns blank, then drop it for the time being
kranji_marshes = GM_buildings.iloc[1293].copy()
kranji_marshes["ESD Consultant"] = '-'
kranji_marshes["ESCO Consultant"] = '-'
kranji_marshes["Green Consultant"] = '-'
GM_buildings = GM_buildings.drop([1293])

# split the ESD column into 3 using split() and add back kranji marshes
GM_buildings[["ESD Consultant", "ESCO Consultant", "Green Consultant"]] = GM_buildings["ESD/ESCO/Green Consultant"].str.split("/", expand=True)

GM_buildings = GM_buildings[:1293].append(kranji_marshes).append(GM_buildings[1293:])
```

```python
# rename columns according to previous dataset
GM_buildings = GM_buildings.rename(columns={"Project Name": "building", "BCA ID": "bca_id", "Project Description:": "description", \
                     "Prominent Green Features:": "green_features", "Award:": "award", "Certification Year:": "fy", \
                     "GFA:": "gfa", "Address:": "address", "Postal Code:": "postal_code", "District:": "district", \
                     "Developer": "developer", "Architect": "architect", "Structural Engineer": "structural_eng", \
                     "M & E Engineer": "me_eng", "Landscape Consultant": "land_cons", "Quality Surveyor": "qs", \
                     "Main Contractor": "main_contractor", "Facility Management": "fac_mgt", "ESD Consultant": "esd_cons", \
                     "ESCO Consultant": "esco_cons", "Green Consultant": "green_cons"})
```

```python
# reorder columns according to previous dataset
GM_buildings = GM_buildings[['bca_id', 'building', 'award', 'fy', 'gfa', 'address', 'postal_code',
       'district', 'developer', 'architect', 'structural_eng', 'me_eng',
       'land_cons', 'qs', 'main_contractor', 'fac_mgt', 'esd_cons',
       'esco_cons', 'green_cons', 'description', 'green_features']]
```

```python
# replace GM award names in the award column
award_names = ['Gold', 'Certified', 'GoldPLUS', 'Platinum', 'Platinum Zero Energy',\
                'Platinum Super Low Energy', 'Platinum Positive Energy', 'GoldPLUS Zero Energy',\
                'Gold Positive Energy', 'GoldPLUS Super Low Energy']
# Gold positive energy & GoldPLUS super low energy are awards not found in the prev dataset

award_names_to_replace = ['Green Mark Gold', 'Green Mark Certified', 'Green Mark Gold Plus', 'Green Mark Platinum',\
                               'Green Mark Platinum (Zero Energy)', 'Green Mark Platinum (Super Low Energy)',\
                               'Green Mark Platinum (Positive Energy)', 'Green Mark Gold Plus (Zero Energy)',\
                              'Green Mark Gold (Positive Energy)', 'Green Mark Gold Plus (Super Low Energy)']
```

```python
GM_buildings['award'] = GM_buildings['award'].replace(award_names, award_names_to_replace)
```

```python
# replace FY names
fy_names = [2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021]

fy_names_to_replace = ['FY06', 'FY07', 'FY08', 'FY09', 'FY10', 'FY11', 'FY12', 'FY13', 'FY14', 'FY15', 'FY16', 'FY17',\
                      'FY18', 'FY19', 'FY20', 'FY21']

GM_buildings["fy"] = GM_buildings["fy"].replace(fy_names, fy_names_to_replace)
```
